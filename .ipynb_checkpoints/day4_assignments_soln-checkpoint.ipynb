```python
import numpy as np
from scipy.interpolate import interp1d
m = np.array([9.9, 10.6, 12.1, 13.8, 15.1, 17.4, 19.4, 21.2])
d = np.array([1620.55, 1639.04, 1673.85, 1704.74, 1713.12, 1697.61, 1660.30, 1618.15])
len(m), len(d)
```

```python
# y = a + bx
# y = p + qx + r x^2
b, a = np.polyfit(m,d,1)
r, q, p = np.polyfit(m,d,2)

def yd1(x):
    return a + b * x
def yd2(x):
    return p + q*x + r * x**2

ysp = interp1d(m,d, kind='quadratic')
```

```python
import matplotlib.pyplot as plt
# drawing fit lines. m = 5 to 25
x = np.linspace(9.9,21.2, 1000)
plt.scatter(m,d)
plt.grid(linestyle='--')
plt.plot(x, yd1(x), color = 'blue')
plt.plot(x, yd2(x), color = 'red')
plt.plot(x, ysp(x), color = 'k', linestyle='--')
```

```python
from scipy.optimize import newton

print("Two roots of quadratic Line of fit are:")
display(newton(yd2, 10) )
display(newton(yd2, 20) )
```
