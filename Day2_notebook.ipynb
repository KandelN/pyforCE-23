# Example 1:

(Example from Book: Fluid Mechanics by Frank M. White ( P 3.140) )


<br> Hint: <br> 
Bernoulli's Equation: $$ P_1 + \frac{1}{2}\rho v_1^2 + \rho g h_1 = P_2 + \frac{1}{2}\rho v_2^2 + \rho g h_2 $$

## Solution:

```python
# Import lib
import sympy as sp

# Assign Symbols:
h2,v2,g = sp.symbols("h_2, v_2, g")

# Initilaize values:
g_val = 9.81    #m/s
h1 = 1.5    #m
H = 4       #m
v1 = 3      #m/s
# h2 = ?    #m ; d/s height

# Apply Bernoulli's equation: (Energy equation)
eq1 = sp.Eq( (H + h1) + v1**2 / (2*g), h2 + v2**2/(2*g) )
display("equation 1:", eq1)
```

```python
# We need to find v2 in terms of v1:

# We know, from question:
# Q1 = Q2
# Q = Av = bh* v
# Assume, b = 1


b = sp.symbols("b")
eq2 = sp.Eq( b * h1 * v1, b * h2 * v2)
display("equation 2:", eq2)
```

```python
eq1.subs(g,9.81)
```

```python
eq2.subs(b,1)
```

```python
ans = sp.solve((eq1.subs(g,g_val), eq2.subs(b,1))  ,  [h2,v2])
ans
```

```python
# for loop in python:

for x in ans:
    display(x[0])
```

## Example: Tabular Computation:
* Print the value of $ \sin(\theta) $ and $ \cos(\theta) $ for $ 0 \leq \theta \leq \pi $.
* Take step size of $ \frac{\pi}{12} $.
* Plot functions $ \sin(\theta) $ and $ \cos(\theta) $ and plot points too.

```python
import numpy as np
```

```python
theta =np.linspace(0, np.pi, 12)
theta
```

```python
for i in range(len(theta)):
    print(f"{theta[i]:.4f}")
```

```python
# Lazy way
theta =np.linspace(0, np.pi, 12)

sin_theta =np.sin(theta)
cos_theta =np.cos(theta)
print(theta, sin_theta, cos_theta)
```

```python
theta =np.linspace(0, np.pi, 12)

sin_theta =np.sin(theta)
cos_theta =np.cos(theta)

print('theta', 'sin', 'cos', sep='\t')
for i in range(len(theta)):
    print(f'{theta[i]:.3f}  {sin_theta[i]:.3f}  {cos_theta[i]:.3f}')
```

# Plotting - Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
plt.grid()

# Plot points
plt.scatter(theta, sin_theta, label=" sin(theta)")
plt.scatter(theta, cos_theta, label=" cos(theta)")

#Let's draw a continuous curve of sine and cosine:
x =np.linspace(-np.pi, 2* np.pi, 1000)
plt.plot(x, np.sin(x), '--')
plt.plot(x, np.cos(x), '--')

plt.legend()
plt.xlabel(" theta")
plt.ylabel("f(x)")
plt.savefig('sin_cos.png', dpi=120)
```

# Simulation Problem
Given a vertical section of hemispherical water tank, simulate waterflow to find the total time it takes to empty the tank when input parameters $H_1$, $H_2$, $R$, $C_d$, $g$ and $dh$ are provided. Symbols have usual meaning. <br>

# Basic Algo:

1. Make function that takes input parameter and outputs t,h lists. ( when h = H1, t = 0 ....).
2. Intilaize variable values
3. call function and get lists of t and h.
4. Plot the values.

```python
import numpy as np
```

```python
def empty_tank(H1, H2, R, a, Cd, g, dh):
    """
    Calculates time elapsed in each step and returns lists t, h.
    """
    K = (np.pi/(Cd * a * (2*g)**0.5))

    #Initilaize variables:
    Time = 0
    h = H1
    
    # Create new list:
    T_list = [0]
    h_list = [H1]
    while (h > H2):
        dt = K * ( 2*R*h**0.5 - h**(3/2)) * dh

        Time = Time + dt
        h = h - dh

        #update list:
        T_list.append(Time)
        h_list.append(h)
        
        # print(f"dh={dh},dt={dt},Time={Time},h={h}")

    print("Total time = ", Time)
    return T_list, h_list
      
```

```python
H1, H2, R, a, Cd = 20, 1, 20, 0.005, 0.6
dh = 0.0001
T_list,h_list =  empty_tank(H1, H2, R, a, Cd, 9.81, dh)
```

```python
# Let's plot this process:
import matplotlib.pyplot as plt
plt.plot(T_list, h_list)
plt.grid()
plt.xlabel('Time elapsed in seconds')
plt.ylabel('Height in meters')
```

```python
def empty_analytically(H1, H2, R, a, Cd, g):
    """
    Calculates time required analytically.
    """
    K = (np.pi/(Cd * a * (2*g)**0.5))
    A = 4/3 * R * ( H1**(3/2) -  H2**(3/2) )
    B = 2/5 * ( H1**(5/2) - H2**(5/2) )

    Time = K*( A - B )
    return Time
empty_analytically(H1, H2, R, a, Cd, 9.81)
```
